```
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```
# Extracting data
data = sns.load_dataset('penguins')
display(data)
```

```
# EDA
print(data.info())
print(data.isnull().sum())
```

```
# Removing null values
data=data.dropna()
print(data.isnull().sum())
```

```
# Converting SEX column to 0 and 1
label = LabelEncoder()
data['sex2'] = label.fit_transform(data['sex'])
print(data['sex2'].unique())
```

```
# Converting ISLAND column to 0 and 1
label = LabelEncoder()
data['island2'] = label.fit_transform(data['island'])
print(data['island2'].unique())
```

```
# Converting SPECIES column to 0 and 1
label = LabelEncoder()
data['species2'] = label.fit_transform(data['species'])
print(data['species2'].unique())
```

```
display(data.isnull().sum())
```

```
# Extracting data
x=data.drop(['sex','species','island','species2'], axis=1)
y=data['species2']

# Split train and test
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=7)

# Model selection
model = RandomForestClassifier(random_state=7)
model.fit(x_train, y_train)

# Check Overfitting
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))

# Model Score
y_pred=model.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred))
print(metrics.r2_score(y_test, y_pred))
```

```
# Predicting for adhoc value
print(model.predict([[40, 12, 120, 6009, 1, 0]]))
```

```
# Enhancing Model
display(pd.DataFrame([model.feature_importances_ * 100], columns=x.columns))
```
